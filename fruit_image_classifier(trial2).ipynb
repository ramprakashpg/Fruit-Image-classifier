```python
import torch
import pandas as pd
from torchvision import transforms, models
from PIL import Image
from torch.autograd import Variable
```

```python
image="4_100.jpg"
```

```python
test_trans = transforms.Compose([transforms.Resize(224), transforms.ToTensor()])
model=torch.load("E:/Stud/AI/dataset/CNN/Fruit-Image-Dataset/fruit_image_classifier(resnet50).pth", map_location=torch.device('cpu'))
model.eval()
```

```python
def predict(image1):
    image=Image.open(image1)
    tensor= test_trans(image).float()
    tensor= tensor.unsqueeze_(0)
    input=Variable(tensor)
    output=model(input)
    index= output.data.cpu().numpy().argmax()
    return index
```

```python
predict(image)
```
